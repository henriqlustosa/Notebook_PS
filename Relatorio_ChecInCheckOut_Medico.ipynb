```python
## Importação das Bibliotecas utilizadas neste Fonte
import requests
import json 
import pandas as pd
```

```python

#Baixa realizadas Pacientes
rel_pac_baixa = requests.get('http://intranethspm:5009/hspmsgh-api/relatorio/baixapacientes/2024-10-01/2024-10-02')
response_baixa_pac = json.loads(rel_pac_baixa.text)
df_baixa_paciente =  pd.DataFrame.from_dict(response_baixa_pac )
```

```python
#Profissionais
profissionais = requests.get('http://intranethspm:5009/hspmsgh-api/relatorio/profissionais/2024-10-01/2024-10-02')
profissionais = json.loads(profissionais.text)
df_profissionais =  pd.DataFrame.from_dict(profissionais )
```

```python
#Extrato realizadas Pacientes
rel_extrato = requests.get('http://intranethspm:5009/hspmsgh-api/relatorio/fichapacientes/2024-10-01/2024-10-02')
response_extrato = json.loads(rel_extrato.text)
df_extrato =  pd.DataFrame.from_dict(response_extrato)
```

```python
df_medico_baixa = pd.merge(df_profissionais, df_baixa_paciente, on="usuario")
df_medico_baixa
```

```python
#Define the array of desired statuses
statuses = ['ALTA MEDICA', 'ALTA COM ENCAMINHAMENTO', 'DESISTENCIA'] # Tirei o tipo EVASAO

# Filter the DataFrame
filtered_df_extrato_baixa = df_extrato[df_extrato['status_ficha'].isin(statuses)]
filtered_df_extrato_baixa
```

```python
df_medico_baixa_especialidade =pd.merge(df_medico_baixa, filtered_df_extrato_baixa,  left_on=["con_numero"], 
    right_on=["cod_consulta"])
df_medico_baixa_especialidade
# Show all columns
pd.set_option('display.max_columns', None)
```
