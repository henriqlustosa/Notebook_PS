```python
import requests
import json 
import pandas as pd
```

```python
# Busca das Fichas de Pacientes Acolhidos pelo endpoint dos relatórios do PS

# EndpointFicha para Paciente Acolhidos
res_ficha_pac_acolhido = requests.get('http://intranethspm:5009/hspmsgh-api/relatorio/fichapacientesacolhidos/2023-03-01/2023-03-31')

```

```python
# Carregamento dos dados passando do formato json para um dicionário de dados

#  Ficha Paciente Acolhidos
response_ficha_pac_acolhido = json.loads(res_ficha_pac_acolhido.text)
```

```python
# Apresentação visual dos dados do dicionário
response_ficha_pac_acolhido
```

```python
# Transformando o tipo de dados: de Dicionário para um Dataframe

# Ficha Paciente Acolhidos
df_ficha_pac_acolhido =  pd.DataFrame.from_dict(response_ficha_pac_acolhido)
```

```python
# Apresentação do DataFrame
df_ficha_pac_acolhido.head()
```

```python
df_ficha_pac_acolhido_novo_indice = df_ficha_pac_acolhido.set_index(['data_acolhimento'])
```

```python
# Convertendo a coluna data_acolhimento to datetimes and adicionando o parâmetro dentor do comando resample:

df_ficha_pac_acolhido['data_acolhimento'] = pd.to_datetime(df_ficha_pac_acolhido['data_acolhimento'])

# Agrupamento do dataFrame por data e hora

serie_ficha_pac_acolhido_produtividade = df_ficha_pac_acolhido.resample('H', on='data_acolhimento').prontuario.count()
```

```python
# Criando um dataframe com a produtividade por hora a partir de uma serie
df_ficha_pac_acolhido_produtividade = serie_ficha_pac_acolhido_produtividade.to_frame()
```

```python
# Apresentação dos dados
df_ficha_pac_acolhido_produtividade.head()
```

```python
# Criado 2 campos para facilitar a busca de informações no dataframe.

df_ficha_pac_acolhido_produtividade['data'] = [d.date() for d in df_ficha_pac_acolhido_produtividade.index]
df_ficha_pac_acolhido_produtividade['hora'] = [d.time() for d in df_ficha_pac_acolhido_produtividade.index]
```
